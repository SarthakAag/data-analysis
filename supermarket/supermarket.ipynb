```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Walmart.csv', encoding_errors='ignore')
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.dtypes
```

```python
df['unit_price'] = df['unit_price'].str.replace('$', '').astype(float)
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df['total'] = df['unit_price'] * df['quantity']
df.head()
```

```python
df[['unit_price', 'quantity', 'total']].head()
```

Most Used Payment Method

```python
plt.figure(figsize=(8,5))
sns.countplot(x='payment_method', data=df)
plt.title('Most Used Payment Method')
plt.show()
```

Sales by Time (Peak Sales Hour)

```python
df['hour'] = pd.to_datetime(df['time']).dt.hour

plt.figure(figsize=(10,5))
sns.countplot(x='hour', data=df)
plt.title('Sales by Hour of Day')
plt.show()
```

Rating Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['rating'], bins=10, kde=True)
plt.title('Customer Ratings Distribution')
plt.show()
```

```python
city_sales = df.groupby('City')['total'].sum().sort_values(ascending=False)
city_sales

```

```python
highest_sales_city = city_sales.idxmax()
highest_sales_value = city_sales.max()

print("City with Highest Sales:", highest_sales_city)
print("Total Sales:", highest_sales_value)
```

```python
lowest_sales_city = city_sales.idxmin()
lowest_sales_value = city_sales.min()

print("City with Lowest Sales:", lowest_sales_city)
print("Total Sales:", lowest_sales_value)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(y='category', x='total', data=df, estimator=sum)
plt.title('Revenue by Product Line')
plt.show()
```

```python
df['date'] = pd.to_datetime(df['date'])
date_sales = df.groupby('date')['total'].sum()


max_sales_date = date_sales.idxmax()
max_sales_value = date_sales.max()

print("Date with Highest Sales:", max_sales_date)
print("Total Sales on that Date:", max_sales_value)
```

```python
min_sales_date = date_sales.idxmin()
min_sales_value = date_sales.min()

print("Date with Lowest Sales:", min_sales_date)
print("Total Sales on that Date:", min_sales_value)
```

Month with Highest Sales

```python
df['month'] = df['date'].dt.month

month_sales = df.groupby('month')['total'].sum()

highest_sales_month = month_sales.idxmax()
highest_month_sales_value = month_sales.max()

print("Month with Highest Sales:", highest_sales_month)
print("Total Sales in that Month:", highest_month_sales_value)
```

Day of Week with Highest Sales

```python
df['day_name'] = df['date'].dt.day_name()

day_sales = df.groupby('day_name')['total'].sum()

highest_sales_day = day_sales.idxmax()
highest_day_sales_value = day_sales.max()

print("Day with Highest Sales:", highest_sales_day)
print("Total Sales on that Day:", highest_day_sales_value)
```

City + Date Combination with Highest Sales

```python
city_date_sales = df.groupby(['City', 'date'])['total'].sum()

highest_city_date = city_date_sales.idxmax()
highest_city_date_sales = city_date_sales.max()

print("City & Date with Highest Sales:", highest_city_date)
print("Sales Amount:", highest_city_date_sales)
```

Sales Trend Over Time

```python
df['date'] = pd.to_datetime(df['date'])

daily_sales = df.groupby('date')['total'].sum()

plt.figure(figsize=(10,5))
daily_sales.plot()
plt.title('Daily Sales Trend')
plt.ylabel('Total Sales')
plt.show()
```
